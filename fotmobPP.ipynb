```python
import LanusStats as ls
import pandas as pd
```

```python
fotmob = ls.FotMob()
```

```python
fotmob.get_match_shotmap(5075634)
```

```python
liga = fotmob.fotmob_request('/data/leagues?id=112&ccode3=ARG')
liga.json()
```

```python
df_liga_arg = pd.DataFrame(liga.json()['fixtures']['allMatches'])
```

```python
df_liga_arg
```

```python
df_liga_arg[df_liga_arg['round'] == '16']
```

```python
df_apertura = df_liga_arg.iloc[:240]
```

```python
ids_partidos = df_apertura.id.unique()
```

```python
from tqdm import tqdm
df_total = pd.DataFrame()
for partido in tqdm(ids_partidos):
    df = fotmob.get_match_shotmap(partido)
    df_total = pd.concat([df_total, df])
```

```python
df_total.situation.value_counts()
```

```python
pelotas_paradas = df_total[df_total['situation'].isin(['FromCorner', 'SetPiece', 'ThrowInSetPiece'])].reset_index(drop=True)
```

```python
pelotas_paradas
```

```python
zonaa = pd.DataFrame(liga.json()['table'][0]['data']['tables'][0]['table']['all'])
zonab = pd.DataFrame(liga.json()['table'][0]['data']['tables'][1]['table']['all'])
ids_fotmob = pd.concat([zonaa,zonab])[['name', 'id']]
```

```python
pelotas_paradas.merge(ids_fotmob, left_on='teamId', right_on='id')[['eventType', 'name', 'playerName', 'x', 'y', 'min', 'expectedGoals', 'expectedGoalsOnTarget', 'shotType','situation','period','id_x']]
```

```python
goles_pelotas_paradas = df_pelotas_paradas[df_pelotas_paradas['eventType'] == 'Goal']
goles_pelotas_paradas = goles_pelotas_paradas.groupby('name').count()['eventType']
```

```python
df_pelotas_paradas.groupby('name').agg({
    'expectedGoals': ['count', 'sum', 'mean']
})
```
